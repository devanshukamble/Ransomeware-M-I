```python
from cryptography import fernet
key = fernet.Fernet.generate_key()
f = fernet.Fernet(key)
while True:
    c = int(input("Enter 1 to encrypt file\nEnter 2 to decrypt file"))
    if c == 1:
        s = ''
        with open('text.txt') as f1:
            print("reading file")
            s = f1.read()
        print("file reading complete")
        s = s.encode("ascii")
        s = f.encrypt(s)
        with open('text.txt','wb') as f1:
            f1.write(s)
        print('file encrypted')
    elif c == 2:
        with open('text.txt','rb') as f1:
            s = f1.read()
        s = f.decrypt(s)
        s = s.decode("ascii")
        with open('text.txt','w') as f1:
            f1.write(s)
    elif c == 3:
        break
```
