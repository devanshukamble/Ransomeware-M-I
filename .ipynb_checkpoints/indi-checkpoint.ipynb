```python
import os
import win32api
drives = win32api.GetLogicalDriveStrings()
drives = drives.split('\000')[:-1]
print(drives)
count = 0
for k in drives:
    for i in os.walk(k):
        for j in i[2]:
            if j.endswith('.txt'):
                count+=1
                print(j,"   ",count)
                print()
print(count)
```
